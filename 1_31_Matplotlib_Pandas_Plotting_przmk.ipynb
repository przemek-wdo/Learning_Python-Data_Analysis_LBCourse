```python
!pip install -U datasets
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## üü® Job Postings Trend Line Plot (1.31.3)

```python
monthly_job_counts = df['job_posted_date'].dt.month.value_counts().sort_index().reset_index()
monthly_job_counts = monthly_job_counts.rename(columns={'count':'Job Postings'})
monthly_job_counts.job_posted_date = pd.to_datetime(monthly_job_counts.job_posted_date, format='%m').dt.strftime('%b')
monthly_job_counts = monthly_job_counts.set_index('job_posted_date')
monthly_job_counts.plot(
    kind='line',
    figsize=(12, 6),
    xticks=range(12),
    title='Monthly Job Postings Trend',
    xlabel='Month',
    ylabel='Number of Job Postings'
)
```
